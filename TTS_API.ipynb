```python
from gtts import gTTS
from IPython.display import Audio
from IPython.display import display
```

```python
text = "이미지에서 얼굴 개인정보가 포함되지 않았습니다."
tts = gTTS(text, lang="ko")
tts.save("error_audio.mp3")
```

```python
wn = Audio('error_audio.mp3', autoplay=True)
display(wn)
```
